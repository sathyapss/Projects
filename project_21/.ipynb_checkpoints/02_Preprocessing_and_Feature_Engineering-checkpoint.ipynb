# AMES Housing DataSet

# Problem Statement:

The second Data Science Project aims to develop the algorith to estimate the price of residential houses based on fixed features i.e. characteristics that cannot be easily renovated (e.g. location, square feet, number of bedrooms and bathrooms)
Identifying characteristics of residential houses that can be cost-effectively renovated and estimating the mean value of these renovations
This project uses the Ames housing data available on Kaggle, which includes 81 features describing a wide range of characteristics of 1,460 homes in Ames, Iowa sold between 2006 and 2010.  Models were required to be trained on houses sold prior to 2010 and evaluated on houses sold in 2010.

What features of a house are most important (have the highest correlation) in predicting its price in the Ames market? Furthermore, can a linear regression model be developed that achieves high accuracy (greater than a baseline RÂ² score of 0.5)?

```python
# Selected the below columns from the train dataset based on the EDA outcome

ntrain=train.loc[:,['MS Zoning','MS SubClass','Lot Area','Lot Shape','Land Contour','Lot Config','Land Slope'
             ,'Overall Qual','Mas Vnr Area','Exter Qual','Exter Cond','Bsmt Qual','Bsmt Cond','Heating QC',
             'Kitchen Qual','Fireplace Qu','Garage Qual','Total Bsmt SF','Gr Liv Area',
             'TotRms AbvGrd','Fireplaces','Garage Area','Bsmt Full Bath','Full Bath',
             'Half Bath','Bedroom AbvGr','Kitchen AbvGr','Pool Area','SalePrice','House Style','Overall Cond','Lot Frontage',
                    'Roof Style','Bsmt Unf SF','Neighborhood']]
```

```python
#Adding relevant features for years Build, Years MOdified & Years Sold
ntrain['years_remod'] =   train['Year Built'] - train['Year Remod/Add'] 
ntrain['years_to_sell'] = train['Yr Sold'] - train['Year Built']
   
#train.loc[:,['years_remod','years_to_sell','Year Built','Year Remod/Add','Yr Sold']]


```

For all the categorical Columns which are not ordinal columns we created the dummies for each vlaue of the column. The columns which falls under this category are as follows:
    Lot Shape, Land Contour, Lot Config, Land Slope, MS Zoning, House Style
    

```python
# Adding Dummies

ntrain = pd.get_dummies(ntrain, columns=['Lot Shape'], prefix = ['Lot Shape'])
ntrain = pd.get_dummies(ntrain, columns=['Land Contour'], prefix = ['Land Contour'])
ntrain = pd.get_dummies(ntrain, columns=['Lot Config'], prefix = ['Lot Config'])
ntrain = pd.get_dummies(ntrain, columns=['Land Slope'], prefix = ['Land Slope'])
ntrain = pd.get_dummies(ntrain, columns=['MS Zoning'], prefix = ['MS Zoning'])
ntrain = pd.get_dummies(ntrain, columns=['House Style'], prefix = ['House Style'])
ntrain = pd.get_dummies(ntrain, columns=['Roof Style'], prefix = ['Roof Style'])
ntrain = pd.get_dummies(ntrain, columns=['Neighborhood'], prefix = ['Neighborhood'])


```

For all the categorical column which are ordinal number which denotes the rating are converted to number by assigning th values.The following columns were converrted to the ordinal number columns.
    Exter Qual, Exterr Cond, Bsmt Qual, Bsmt Cond , Heating QC,Kitchen Qual,Fireplace Qu,Garage Qual

```python
# Adding Ordinal Values

ord_num = {'Ex': 5, 
           'Gd': 4, 
           'TA': 3,
           'Fa': 2,
           'Po': 1, 
           'NA': 0}

ntrain['Exter Qual Ord'] = ntrain['Exter Qual'].replace(ord_num)
ntrain['Exter Cond Ord'] = ntrain['Exter Cond'].replace(ord_num)
ntrain['Bsmt Qual Ord'] = ntrain['Bsmt Qual'].replace(ord_num)
ntrain['Bsmt Cond Ord'] = ntrain['Bsmt Cond'].replace(ord_num)
ntrain['Heating QC Ord'] = ntrain['Heating QC'].replace(ord_num)
ntrain['Kitchen Qual Ord'] = ntrain['Kitchen Qual'].replace(ord_num)
ntrain['Fireplace Qu Ord'] = ntrain['Fireplace Qu'].replace(ord_num)
ntrain['Garage Qual Ord'] = ntrain['Garage Qual'].replace(ord_num)

```

```python
#Adding Zero when 'Mas Vnr Areaâ is null
ntrain = ntrain.fillna(train['Mas Vnr Area'].value_counts().index[0])
ntrain = ntrain.fillna(train['Lot Frontage'].value_counts().index[0])
```

New Predictor PorchArea is added by summing all the values of the following columns
    'Open Porch SF' , 'Enclosed Porch','3Ssn Porch','Screen Porch' 
    
    Dropped all the above four columns

```python
# Add Column Porch Area adding the sum of Open Porch SF	Enclosed Porch	3Ssn Porch	Screen Porch

ntrain['PorchArea'] = train['Open Porch SF']+train['Enclosed Porch']+train['3Ssn Porch']+train['Screen Porch']
```

```python
# Drop columns  for ordinal number
ntrain.drop(columns=['Exter Qual', 'Exter Cond','Bsmt Qual','Bsmt Cond','Heating QC','Kitchen Qual','Fireplace Qu',
                    'Garage Qual'],inplace=True)
```

```python
ntrain.isnull().sum()
```

# ANOVA for Feature Selection

```python
features = [col for col in ntrain.columns if col != 'SalePrice']
X = ntrain[features]
y = ntrain['SalePrice']

ss = StandardScaler()
ss.fit(X)
X_sc = ss.transform(X)
```

```python
# Create an SelectKBest object to select features with two best ANOVA F-Values
fvalue_selector = SelectKBest(f_classif, k=20)

# Apply the SelectKBest object to the features and target
X_kbest = fvalue_selector.fit_transform(X_sc, y)
```

```python
# Show results
print('Original number of features:', X.shape)
print('Reduced number of features:', X_kbest.shape)

#gdsc.feature_names
```

# Train Test Split / Standard Scalar

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```
