# AMES Housing DataSet

# Problem Statement:

The second Data Science Project aims to develop the algorithm to estimate the price of residential houses based on fixed features i.e. characteristics that cannot be easily renovated (e.g. location, square feet, number of bedrooms and bathrooms)

This project uses the Ames housing data available on Kaggle, which includes 81 features describing a wide range of characteristics of 1,460 homes in Ames, Iowa sold between 2006 and 2010.  

What features of a house are most important (have the highest correlation) in predicting its price in the Ames market?

# Feature Selection - First Cut

Based on the EDA outcomes, we select the features and assign the dumminies , ordinal number for the feature. 

```python
# Selected the below columns from the train dataset based on the EDA outcome

ntrain=train.loc[:,['MS Zoning','MS SubClass','Lot Area','Lot Shape','Land Contour','Lot Config','Land Slope'
             ,'Overall Qual','Mas Vnr Area','Exter Qual','Exter Cond','Bsmt Qual','Bsmt Cond','Heating QC',
             'Kitchen Qual','Fireplace Qu','Garage Qual','Total Bsmt SF','Gr Liv Area',
             'TotRms AbvGrd','Fireplaces','Garage Area','Bsmt Full Bath','Full Bath',
             'Half Bath','Bedroom AbvGr','Kitchen AbvGr','Pool Area','SalePrice','House Style','Overall Cond','Lot Frontage',
                    'Roof Style','Bsmt Unf SF','Neighborhood']]
```

```python
#Adding relevant features for years Build, Years MOdified & Years Sold
ntrain['years_remod'] =   train['Year Built'] - train['Year Remod/Add'] 
ntrain['years_to_sell'] = train['Yr Sold'] - train['Year Built']
   
#train.loc[:,['years_remod','years_to_sell','Year Built','Year Remod/Add','Yr Sold']]


```

For all the categorical Columns which are not ordinal columns we created the dummies for each vlaue of the column. The columns which falls under this category are as follows:
    Lot Shape, Land Contour, Lot Config, Land Slope, MS Zoning, House Style
    

```python
# Adding Dummies

ntrain = pd.get_dummies(ntrain, columns=['Lot Shape'], prefix = ['Lot Shape'])
ntrain = pd.get_dummies(ntrain, columns=['Land Contour'], prefix = ['Land Contour'])
ntrain = pd.get_dummies(ntrain, columns=['Lot Config'], prefix = ['Lot Config'])
ntrain = pd.get_dummies(ntrain, columns=['Land Slope'], prefix = ['Land Slope'])
ntrain = pd.get_dummies(ntrain, columns=['MS Zoning'], prefix = ['MS Zoning'])
ntrain = pd.get_dummies(ntrain, columns=['House Style'], prefix = ['House Style'])
ntrain = pd.get_dummies(ntrain, columns=['Roof Style'], prefix = ['Roof Style'])
ntrain = pd.get_dummies(ntrain, columns=['Neighborhood'], prefix = ['Neighborhood'])


```

For all the categorical column which are ordinal number which denotes the rating are converted to number by assigning th values.The following columns were converrted to the ordinal number columns.
    Exter Qual, Exterr Cond, Bsmt Qual, Bsmt Cond , Heating QC,Kitchen Qual,Fireplace Qu,Garage Qual

```python
# Adding Ordinal Values

ord_num = {'Ex': 5, 
           'Gd': 4, 
           'TA': 3,
           'Fa': 2,
           'Po': 1, 
           'NA': 0}

ntrain['Exter Qual Ord'] = ntrain['Exter Qual'].replace(ord_num)
ntrain['Exter Cond Ord'] = ntrain['Exter Cond'].replace(ord_num)
ntrain['Bsmt Qual Ord'] = ntrain['Bsmt Qual'].replace(ord_num)
ntrain['Bsmt Cond Ord'] = ntrain['Bsmt Cond'].replace(ord_num)
ntrain['Heating QC Ord'] = ntrain['Heating QC'].replace(ord_num)
ntrain['Kitchen Qual Ord'] = ntrain['Kitchen Qual'].replace(ord_num)
ntrain['Fireplace Qu Ord'] = ntrain['Fireplace Qu'].replace(ord_num)
ntrain['Garage Qual Ord'] = ntrain['Garage Qual'].replace(ord_num)

```

New Predictor PorchArea is added by summing all the values of the following columns
    'Open Porch SF' , 'Enclosed Porch','3Ssn Porch','Screen Porch' 
    
    Dropped all the above four columns

```python
# Add Column Porch Area adding the sum of Open Porch SF	Enclosed Porch	3Ssn Porch	Screen Porch

ntrain['PorchArea'] = train['Open Porch SF']+train['Enclosed Porch']+train['3Ssn Porch']+train['Screen Porch']
```

```python
# Drop columns  for ordinal number
ntrain.drop(columns=['Exter Qual', 'Exter Cond','Bsmt Qual','Bsmt Cond','Heating QC','Kitchen Qual','Fireplace Qu',
                    'Garage Qual'],inplace=True)
```

```python
#Adding Zero when feature is null
ntrain['Mas Vnr Area'] = ntrain['Mas Vnr Area'].fillna(0).astype('float64')
ntrain['Fireplace Qu Ord'] = ntrain['Fireplace Qu Ord'].fillna(0).astype('float64')
ntrain['Garage Qual Ord'] = ntrain['Garage Qual Ord'].fillna(0).astype('float64')
ntrain['Total Bsmt SF'] = ntrain['Total Bsmt SF'].fillna(0).astype('float64')
ntrain['Garage Area'] = ntrain['Garage Area'].fillna(0).astype('float64')
ntrain['Bsmt Full Bath'] = ntrain['Bsmt Full Bath'].fillna(0).astype('float64')
ntrain['Lot Frontage'] = ntrain['Lot Frontage'].fillna(0).astype('float64')
ntrain['Bsmt Unf SF'] = ntrain['Bsmt Unf SF'].fillna(0).astype('float64')
ntrain['Bsmt Qual Ord'] = ntrain['Bsmt Qual Ord'].fillna(0).astype('float64')
ntrain['Bsmt Cond Ord'] = ntrain['Bsmt Cond Ord'].fillna(0).astype('float64')
```

```python
ntrain['Heating QC Ord'] =ntrain['Heating QC Ord'].astype('int64')
ntrain['Exter Qual Ord'] =ntrain['Exter Qual Ord'].astype('int64')
ntrain['Kitchen Qual Ord'] =ntrain['Kitchen Qual Ord'].astype('int64')
```

# Train Test Split / Standard Scalar

```python
features = [col for col in ntrain.columns if col != 'SalePrice']
X = ntrain[features]
y = ntrain['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ss = StandardScaler()
ss.fit(X_train)
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

```python
X.shape
```

# Model Preparation # Model1

The first model we tried with linear regreassion considering all the 94 features. But the R2 score is in negative and RMSE also very high 3476171755886500.0

```python
lr = LinearRegression()
lr_scores = cross_val_score(lr, X_train, y_train, cv=5)
lr_scores.mean() 
```

```python
#Fit the model and find the score
lr.fit(X_train,y_train)
print("Score for train data Model1(Linear Regression):   " + str (lr.score(X_train, y_train)))
print("Score for test data Model1(Linear Regression):   " + str (lr.score(X_test, y_test)))

# predict the y values based on the model1
y_predict=lr.predict(X_test)
m1=mean_squared_error(y_test,y_predict)
root_mean_squared_error = sqrt(m1)
print("Root Mean Square Error:   " + str(root_mean_squared_error))
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_predict)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title('Plot of Actaul versus Predicted test split data for Linear Regression')
plt.show()
```

# Model2  --  Lasso

The model 2 Lasso is initialized and validated against the test data fromtrain/test split

```python
#INitialize the lasso and find the cross validation score
lasso = LassoCV(n_alphas=200)
lasso_scores = cross_val_score(lasso, X_train, y_train, cv=5)
lasso_scores.mean()
```

```python
lasso.fit(X_train,y_train)
print("Score for Train Data Model2(Lasso):   " + str(lasso.score(X_train, y_train)))
print("Score for test Data Model2(Lasso):    " + str(lasso.score(X_test, y_test)))
y_predict=lasso.predict(X_test)

m1=mean_squared_error(y_test,y_predict)
root_mean_squared_error = sqrt(m1)
print("RMSE for the test data:    " + str(root_mean_squared_error))
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_predict)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title('Plot of Actaul versus Predicted test split data for Lasso Regression')
plt.show()
```

# Removing Features Based on Lasso

```python
removed_feats = X.columns[(lasso.coef_ == 0).ravel().tolist()]
```

# Model3 Ridge

```python
features = [col for col in ntrain.columns if col != 'SalePrice']
X = ntrain[features]
y = ntrain['SalePrice']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ss = StandardScaler()
ss.fit(X_train)
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

```python
ridge = RidgeCV(alphas=np.linspace(.1, 10, 100))
ridge_scores = cross_val_score(ridge, X_train, y_train, cv=5)
ridge_scores.mean()
```

```python
ridge.fit(X_train,y_train)
print("Score for train data Model3 Ridge:   " + str(ridge.score(X_train, y_train)))
print("score for test data Model3 Ridge:   "+ str(ridge.score(X_test, y_test)))

y_predict=ridge.predict(X_test)
m1=mean_squared_error(y_test,y_predict)
root_mean_squared_error = sqrt(m1)
print("RMSE :     " + str(root_mean_squared_error))
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_predict)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Actual')
ax.set_ylabel('Predicted')
ax.set_title('Plot of Actaul versus Predicted test split data for Ridge Regression')
plt.show()
```

```python
pd.Series(ridge.coef_, index=features).plot.barh(figsize=(15, 15))
```
