# Reading the data

```python
# Importing Necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from math import sqrt


from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV , ElasticNet
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# Reading Data from the files
train = pd.read_csv('./datasets/train.csv')
test = pd.read_csv('./datasets/test.csv')
```

# EDA

# 01: Visualizing the Sales Price:

The Sales price has right skewed distribution. The boxplot also intrepects that the outliers are above the maximum line (Q3+1.5IQR)

```python
n, bins, patches = plt.hist(x=train['SalePrice'], bins=30, color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Sales Price')
plt.ylabel('Frequency')
plt.title('Sales Price of plot')
plt.show()

boxplot = train.boxplot(column=['SalePrice'])
```

# 02: Utilities/Street Count of records by value

All the records have same value "AllPub" , so this feature will not help to predict the sale price. And we can elimate from the features for predicting the sale price.

Similarly the street has same value for 2044 records, so we can elimate the street from the feature selection.

```python
ax = train['Utilities'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Number for each Utilities")
ax.set_xlabel("Utilities")
ax.set_ylabel("Frequency")

train['Utilities'].value_counts()
```

```python
ax = train['Street'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Number for each Street")
ax.set_xlabel("Street")
ax.set_ylabel("Frequency")

train['Street'].value_counts()
```

# 03. Neighborhood Visualization

```python
ax = train['Neighborhood'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Number for each Neighborhood")
ax.set_xlabel("Neighborhood]")
ax.set_ylabel("Frequency")

```

# 04.Overall Quality 

The Overall quality is useful features which gives ordinal ranking  on the quality of the house sold.

```python
ax = train['Overall Qual'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Number for each Overall Quality")
ax.set_xlabel("Overall Qual")
ax.set_ylabel("Frequency")

train['Overall Qual'].value_counts()
```

# 05. Land Slope 

The land Slope column has most of the values are Gtl

```python
ax = train['Land Slope'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Number for each Land Slope")
ax.set_xlabel("Land Slope")
ax.set_ylabel("Frequency")

train['Land Slope'].value_counts()
```

# 05. Sales Price versus Years Built

```python
ax1 = train.plot.scatter(x='Year Built',y='SalePrice',c='Blue')
w1=train.groupby('Year Built')['SalePrice'].mean()

ax2 = w1.plot.line(x='Year Built',y='SalePrice',color='Red',label='Mean Sales Price per year', linestyle='--')

ax1.set_title('Sales Price Versus Years Build')
ax1.legend()
```

# 06. Sales Price versus Years Mod

```python
ax1 = train.plot.scatter(x='Year Remod/Add',y='SalePrice',c='Blue')

w1=train.groupby('Year Remod/Add')['SalePrice'].mean()

ax2 = w1.plot.line(x='Year Remod/Add',y='SalePrice',color='Red',label='Mean Sales Price per year Mod', linestyle='--')

ax1.set_title('Sales Price Versus Years Remod')
ax1.legend()
```
