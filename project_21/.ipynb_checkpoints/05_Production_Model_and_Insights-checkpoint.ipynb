# AMES Housing DataSet

# Problem Statement:

The second Data Science Project aims to develop the algorith to estimate the price of residential houses based on fixed features i.e. characteristics that cannot be easily renovated (e.g. location, square feet, number of bedrooms and bathrooms)
Identifying characteristics of residential houses that can be cost-effectively renovated and estimating the mean value of these renovations
This project uses the Ames housing data available on Kaggle, which includes 81 features describing a wide range of characteristics of 1,460 homes in Ames, Iowa sold between 2006 and 2010.  Models were required to be trained on houses sold prior to 2010 and evaluated on houses sold in 2010.

What features of a house are most important (have the highest correlation) in predicting its price in the Ames market? Furthermore, can a linear regression model be developed that achieves high accuracy (greater than a baseline R² score of 0.5)?

# Reading the data

```python
# Importing Necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from math import sqrt


from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression, LassoCV, RidgeCV , ElasticNet
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score, cross_val_predict
from sklearn.metrics import r2_score,mean_squared_error
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
# Reading Data from the files
ntrain = pd.read_csv('./datasets/ntrain_features.csv')
test = pd.read_csv('./datasets/test.csv')
```

# Model Fitting -Ridge

I tried to improve the model performance by grid searching on the ElasticNet and LassoCV hyperparameters, and removed some of features .  Knowing that Lasso regularisation results in a sparser model and deals with multicollinearity between predictors well, I decided to choose the optimum Ridge model  for predicting the sale price of residential houses based on fixed features.  This model achieved a very similar R2 score (0.869) to the linear regression model above – but importantly, I removed 6 predictors.  The 20 largest model coefficients in terms of absolute value, are displayed in the bar plot below.

```python
ntrain.shape
```

```python
features = [col for col in ntrain._get_numeric_data().columns if col != 'SalePrice']
X = ntrain[features]
y = ntrain['SalePrice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
ss = StandardScaler()
ss.fit(X_train)
X_train = ss.transform(X_train)
X_test = ss.transform(X_test)
```

```python
X.shape
```

```python
ridge = RidgeCV(alphas=np.linspace(.1, 10, 100))
```

```python
ridge.fit(X_train, y_train)
```

```python
ridge.score(X_train, y_train)
```

```python
pred = ridge.predict(X_test)
```

```python
r2_score(y_test, pred)
```

```python
m1=mean_squared_error(y_test,pred)
root_mean_squared_error = sqrt(m1)
print(root_mean_squared_error)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, pred)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Actual Sale Price')
ax.set_ylabel('Predicted Sale Price')
ax.set_title('Plot of Actual versus Predicted test split data for Rigde  Regression')
plt.show()
```

```python
pd.Series(ridge.coef_, index=features).plot.barh(figsize=(15, 7))
plt.title("20 Most Significant predictors based on Ridge Regression")
```

# Test Dataset Cleaning & SalePrice Prediction

```python
# Selected the required columns from the train dataset based on the EDA outcome

ntest=test.loc[:,['MS Zoning','MS SubClass','Lot Area','Lot Shape','Land Contour','Lot Config','Land Slope'
             ,'Overall Qual','Mas Vnr Area','Exter Qual','Exter Cond','Bsmt Qual','Bsmt Cond','Heating QC',
             'Kitchen Qual','Fireplace Qu','Garage Qual','Total Bsmt SF','Gr Liv Area',
             'TotRms AbvGrd','Fireplaces','Garage Area','Bsmt Full Bath','Full Bath',
             'Half Bath','Bedroom AbvGr','Kitchen AbvGr','Pool Area','House Style','Overall Cond','Lot Frontage',
                    'Roof Style','Bsmt Unf SF','Neighborhood']]
```

```python
#Adding relevant features for years Build, Years MOdified & Years Sold
ntest['years_remod'] =   test['Year Built'] - test['Year Remod/Add'] 
ntest['years_to_sell'] = test['Yr Sold'] - test['Year Built']
   
#train.loc[:,['years_remod','years_to_sell','Year Built','Year Remod/Add','Yr Sold']]
```

```python
#Adding Zero when 'Mas Vnr Area’ is null
ntest = ntest.fillna(ntest['Mas Vnr Area'].value_counts().index[0])
```

```python
# Adding Dummies

ntest = pd.get_dummies(ntest, columns=['Lot Shape'], prefix = ['Lot Shape'])
ntest = pd.get_dummies(ntest, columns=['Land Contour'], prefix = ['Land Contour'])
ntest = pd.get_dummies(ntest, columns=['Lot Config'], prefix = ['Lot Config'])
ntest = pd.get_dummies(ntest, columns=['Land Slope'], prefix = ['Land Slope'])
ntest = pd.get_dummies(ntest, columns=['MS Zoning'], prefix = ['MS Zoning'])
ntest = pd.get_dummies(ntest, columns=['House Style'], prefix = ['House Style'])
ntest = pd.get_dummies(ntest, columns=['Roof Style'], prefix = ['Roof Style'])
ntest = pd.get_dummies(ntest, columns=['Neighborhood'], prefix = ['Neighborhood'])
```

```python
# Adding Ordinal Values

ord_num = {'Ex': 5, 
           'Gd': 4, 
           'TA': 3,
           'Fa': 2,
           'Po': 1, 
           'NA': 0}

ntest['Exter Qual Ord'] = ntest['Exter Qual'].replace(ord_num)
ntest['Exter Cond Ord'] = ntest['Exter Cond'].replace(ord_num)
ntest['Bsmt Qual Ord'] = ntest['Bsmt Qual'].replace(ord_num)
ntest['Bsmt Cond Ord'] = ntest['Bsmt Cond'].replace(ord_num)
ntest['Heating QC Ord'] = ntest['Heating QC'].replace(ord_num)
ntest['Kitchen Qual Ord'] = ntest['Kitchen Qual'].replace(ord_num)
ntest['Fireplace Qu Ord'] = ntest['Fireplace Qu'].replace(ord_num)
ntest['Garage Qual Ord'] = ntest['Garage Qual'].replace(ord_num)

```

```python
# Add Column Porch Area adding the sum of Open Porch SF	Enclosed Porch	3Ssn Porch	Screen Porch

ntest['PorchArea'] = test['Open Porch SF']+test['Enclosed Porch']+test['3Ssn Porch']+test['Screen Porch']
```

```python
ntest.drop(columns=list(set(ntest.columns)-set(ntrain.columns)),inplace=True)

```

```python
# Standardize the scalar

ss = StandardScaler()
ss.fit(ntest)
X_ntest = ss.transform(ntest)
```

```python
list(set(ntest.columns)-set(ntrain.columns))
```

```python
ntest.shape
```

```python
y_pred = ridge.predict(X_ntest)
```

```python
test['SalePrice'] = y_pred
```

```python
n, bins, patches = plt.hist(x=test['SalePrice'], bins=30, color='Red',
                            alpha=0.7, rwidth=0.85)
n, bins, patches = plt.hist(x=ntrain['SalePrice'], bins=30, color='Blue',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Sales Price')
plt.ylabel('Frequency')
plt.title('Sales Price of plot Predicted for Test Againt the model build on Train Dataset')
plt.show()
```

```python
test_prediction=test.loc[:,['Id','SalePrice']]
```

```python
test_prediction.to_csv('./datasets/ridge_submission.csv',index=False)
```

# Model Fitting -Lasso

I tried to improve the model performance by grid searching on the ElasticNet and LassoCV hyperparameters, and removed some of features .  Knowing that Lasso regularisation results in a sparser model and deals with multicollinearity between predictors well, I decided to choose the optimum Ridge model  for predicting the sale price of residential houses based on fixed features.  This model achieved a very similar R2 score (0.869) to the linear regression model above – but importantly, I removed 6 predictors.  The 20 largest model coefficients in terms of absolute value, are displayed in the bar plot below.

```python
lasso = LassoCV(n_alphas=200)
```

```python
lasso.fit(X_train, y_train)
```

```python
lasso.score(X_train, y_train)
```

```python
pred = lasso.predict(X_test)
```

```python
r2_score(y_test, pred)
```

```python
m1=mean_squared_error(y_test,pred)
root_mean_squared_error = sqrt(m1)
print(root_mean_squared_error)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, pred)
ax.plot([y.min(), y.max()], [y.min(), y.max()], 'k--', lw=4)
ax.set_xlabel('Actual Sale Price')
ax.set_ylabel('Predicted Sale Price')
ax.set_title('Plot of Actual versus Predicted test split data for Rigde  Regression')
plt.show()
```

```python
pd.Series(lasso.coef_, index=features).plot.barh(figsize=(15, 7))
plt.title("20 Most Significant predictors based on lasso Regression")
```
